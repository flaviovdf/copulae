```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
import glob

for f_train, f_tst in zip(glob.glob('*/*trn*'), glob.glob('*/*tst*')):    
    df_trn = pd.read_csv(f_train, header=None)
    df_tst = pd.read_csv(f_tst, header=None)

    if df_trn.shape[1] == 2:
        fig, ax = plt.subplots(2, 1, figsize=(4, 2.5))
        df_trn[0].plot.kde(ax=ax[0], label='Train Dim 0')
        df_tst[0].plot.kde(ax=ax[0], label='Test Dim 0')
        df_trn[1].plot.kde(ax=ax[1], label='Train Dim 1')
        df_tst[1].plot.kde(ax=ax[1], label='Test Dim 1')
        ax[0].set(title=f_train.split('.')[0])
        ax[1].set(title=f_train.split('.')[0])
        ax[0].legend()
        ax[1].legend()
    else:
        fig, ax = plt.subplots(1, 1, figsize=(4, 2.5))
        df_trn[0].plot.kde(ax=ax, label='Train')
        df_tst[0].plot.kde(ax=ax, label='Test')
        ax.set(title=f_train.split('_')[0])
        ax.legend()
    fig.tight_layout(pad=0)
    plt.show()
```
